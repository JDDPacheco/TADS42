```python
path = 'D:/Salve_Aqui_Seus_Arquivos/DIOGO/TADS42/TETI/Arquivos para atividades/'
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
df_mun = pd.read_csv(path+'municipios.csv')
```

```python
df_mun.head()
```

```python
geometry = gpd.points_from_xy(df_mun.longitude, df_mun.latitude)
```

```python
gdf = gpd.GeoDataFrame(df_mun, geometry=geometry)
```

```python
gdf.plot(marker='o', color='green', markersize=5)
```

```python
df_mun2 = pd.read_csv(path+'municipios_2.csv',sep = ';')
```

```python
df_mun2.head()
```

```python
gdf2 = gpd.GeoDataFrame(df_mun2, geometry=geometry)
```

```python
gdf2.plot(marker='s', color='darkblue', markersize=10)
```

```python
brasil = gpd.read_file(path + 'ne_110m_admin_0_countries.shp')
brasil = brasil[brasil.NAME == 'Brazil']
```

```python
fig, ax = plt.subplots(figsize = (10 , 10))
brasil.plot(ax = ax, color = 'gold', edgecolor = 'red')
gdf.plot(ax = ax, marker = 'o', color = 'darkgreen', markersize = 30, alpha = 0.5 ,edgecolor = 'k')
plt.title('Municípios Brasileiros')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
