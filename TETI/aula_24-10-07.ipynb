# Line Plot

# Bar Plot

```python
bar = sns.barplot(x = 'Mes_ID', y = 'Vendas',  data = df_mkt)

df_mkt[['Mes_ID', 'Vendas']].groupby('Mes_ID', as_index = False).agg({'Vendas':'mean'})
```
