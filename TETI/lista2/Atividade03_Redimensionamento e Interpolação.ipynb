```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
from pylab import *
```

```python
img_1 = cv2.imread(path +'image1.jpeg')
img_2 = cv2.imread(path +'image2.jpeg')
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
```

```python
def plot_im2 (im1, tit1, im2, tit2, tam):
    figure(figsize = tam)
    plt.subplot(1, 2, 1), plt.imshow(im1), plt.title(tit1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(1, 2, 2), plt.imshow(im2), plt.title(tit2)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
def plot_im3(im1, tit1, im2, tit2, im3, tit3, tam):
    figure(figsize = tam)
    plt.subplot(1,3,1), plt.imshow(im1), plt.title(tit1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(1,3,2), plt.imshow(im2), plt.title(tit2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(1,3,3), plt.imshow(im3), plt.title(tit3)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
plot_im2(image1,"1", image2, "2", (15,12))
```

## conversão da img_1

```python
img1_30_cubic = cv2.resize(img_1, (0, 0), fx = 0.3, fy = 0.3, interpolation = cv2.INTER_CUBIC)
img1_15_cubic = cv2.resize(img_1, (0, 0), fx = 0.15, fy = 0.15, interpolation = cv2.INTER_CUBIC)
img1_30_area = cv2.resize(img_1, (0, 0), fx = 0.3, fy = 0.3, interpolation = cv2.INTER_AREA)
img1_15_area= cv2.resize(img_1, (0, 0), fx = 0.15, fy = 0.15, interpolation = cv2.INTER_AREA)
img1_30_nearest = cv2.resize(img_1, (0, 0), fx = 0.3, fy = 0.3, interpolation = cv2.INTER_NEAREST)
img1_15_nearest= cv2.resize(img_1, (0, 0), fx = 0.15, fy = 0.15, interpolation = cv2.INTER_NEAREST)
img1_30_l = cv2.resize(img_1, (0, 0), fx = 0.3, fy = 0.3, interpolation = cv2.INTER_LANCZOS4)
img1_15_l= cv2.resize(img_1, (0, 0), fx = 0.15, fy = 0.15, interpolation = cv2.INTER_LANCZOS4)
```

## Conversão da img_2

```python
img_2_30_cubic = cv2.resize(img_2, (0, 0), fx = 0.3, fy = 0.3, interpolation = cv2.INTER_CUBIC)
img_2_15_cubic = cv2.resize(img_2, (0, 0), fx = 0.15, fy = 0.15, interpolation = cv2.INTER_CUBIC)
img_2_30_area = cv2.resize(img_2, (0, 0), fx = 0.3, fy = 0.3, interpolation = cv2.INTER_AREA)
img_2_15_area= cv2.resize(img_2, (0, 0), fx = 0.15, fy = 0.15, interpolation = cv2.INTER_AREA)
img_2_30_nearest = cv2.resize(img_2, (0, 0), fx = 0.3, fy = 0.3, interpolation = cv2.INTER_NEAREST)
img_2_15_nearest= cv2.resize(img_2, (0, 0), fx = 0.15, fy = 0.15, interpolation = cv2.INTER_NEAREST)
img_2_30_l = cv2.resize(img_2, (0, 0), fx = 0.3, fy = 0.3, interpolation = cv2.INTER_LANCZOS4)
img_2_15_l= cv2.resize(img_2, (0, 0), fx = 0.15, fy = 0.15, interpolation = cv2.INTER_LANCZOS4)
```
